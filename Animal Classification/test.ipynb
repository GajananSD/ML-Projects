```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import load_model
```

```python
model = load_model('model.h5')
```

```python
class_names = ['Bear', 'Bird', 'Cat', 'Cow', 'Deer', 'Dog', 'Dolphin', 'Elephant', 'Giraffe', 'Horse', 'Kangaroo', 'Lion', 'Panda',
               'Tiger', 'Zebra']
IMG_SIZE = (224, 224)
```

```python
def predict_animal(image_path):

    img = tf.keras.utils.load_img(image_path, target_size=IMG_SIZE)
    img_array = tf.keras.utils.img_to_array(img) / 255.0
    img_array = tf.expand_dims(img_array, axis=0)  

    pred = model.predict(img_array)
    predicted_class = class_names[np.argmax(pred[0])]
    confidence = np.max(pred[0])

    plt.imshow(img)
    plt.title(f"Predicted: {predicted_class}\nConfidence: {confidence:.2%}")
    plt.axis('off')
    plt.show()

    return predicted_class, confidence
```

```python
import os

folder_path = 'test_images/'

for filename in os.listdir(folder_path):
    if filename.lower().endswith(('.jpeg')):  
        image_path = os.path.join(folder_path, filename)
        animal, conf = predict_animal(image_path)
        print(f"{filename}: Prediction: {animal} ({conf:.2%} confidence)")
```

## 📂 Image Upload and Prediction Script

```python
from tkinter import filedialog
from tkinter import Tk

def upload_image():
    root = Tk()
    root.withdraw()  
    file_path = filedialog.askopenfilename()
    return file_path
```

```python
image_path = upload_image()
if image_path:
    animal, conf = predict_animal(image_path)
    print(f"Prediction: {animal} ({conf:.2%} confidence)")
else:
    print("No file selected")
```
